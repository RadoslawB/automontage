```python
def mp3_to_ndarray(input_path):
    from scipy.io import wavfile
    from pydub import AudioSegment
    
    tmp_path = '__tmp__.wav'
    AudioSegment\
    .from_mp3(input_path)\
    .export(tmp_path, format='wav')

    rv =  wavfile.read(tmp_path) # (rate, ndarray)
    
    os.remove(tmp_path)
    
    return rv
```

```python
rate, data = mp3_to_ndarray('../data/bensound-allthat.mp3')
```

```python
def show_wave(data):
    x = np.arange(0, data.shape[0])
    ax = plt.axes()
    ax.axis('off')
    return ax.plot(x,data, linewidth=.4)
```
